```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

health_data = pd.read_csv('Life Expectancy Data.csv')

print(health_data.head())
```

```python
import re

columns = health_data.keys()
cleaned_columns = [
    re.sub(r'[^0-9A-Za-z]+', '_', col.strip()).lower()
    for col in columns
]
health_data.columns = cleaned_columns

sns.scatterplot(x='schooling', y='life_expectancy', hue='status', palette='bright', data=health_data)
plt.show()
```
